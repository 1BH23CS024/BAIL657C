```python
import os
from langchain_community.document_loaders import TextLoader
from langchain_core.prompts import PromptTemplate
from langchain_cohere import ChatCohere
```

```python
os.environ["COHERE_API_KEY"] = "YOUR_COHERE_API_KEY"
# EXAMPLE: os.environ["COHERE_API_KEY"] = "ghUIJopNUmxOZMyaLdtlOpXD4HCNQOOZbGSEFwFE"
# Note: The above is fake and does not work. Generate your own at https://cohere.com

file_path = "FILE_PATH_TO_INPUT_TEXT"
# EXAMPLE: file_path = "/content/drive/MyDrive/industrial_revolution.txt"
# Step 1: Go to 'Files' tab from the sidebar in Colab
# Step 2: Mount Google Drive by clicking the folder icon (3rd from left)
```

```python
text = TextLoader(file_path).load()[0].page_content
chain = PromptTemplate.from_template(f"Summarize:\n{text}") | ChatCohere()

print(chain.invoke({"text": text}).content)
```
