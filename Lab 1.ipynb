```python
import gensim.downloader as api
from scipy.spatial.distance import cosine
```

```python
model = api.load("glove-wiki-gigaword-100")
```

```python
print(f"First 10 dimensions of 'king' word vector: {model["king"][:10]}")
```

```python
def find_analogy(positive, negative, topn):
  similar_words = model.most_similar(positive, negative, topn)

  for word, similarity in similar_words:
    print(f"{word} (similarity: {similarity:.4f})")
```

```python
print("king + woman - 'man' ≈ ?")
find_analogy(["king", "woman"], "man", 1)
```

```python
print("paris + italy - france ≈ ?\n")
find_analogy(["paris", "italy"], "france", 5)
```

```python
similarity = 1 - cosine(model["king"], model["queen"])
print(f"Cosine similarity between 'king' and 'queen' is {similarity:.4f}")
```
