```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
file_name = "/content/drive/MyDrive/IPC.pdf"
```

```python
vectorstore = FAISS.from_documents(
    CharacterTextSplitter(chunk_size=1000, chunk_overlap=100).split_documents(PyPDFLoader(file_name).load()),
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
)
```

```python
def ipc_chatbot(q):
    for i, doc in enumerate(vectorstore.similarity_search(q, k=3), 1):
        print(f"\n{i}. {doc.page_content[:500]}\n" + "-" * 50)
```

```python
print("Type 'exit' to quit\n")
while (q := input("YOU: ").strip()).lower() != "exit":
    ipc_chatbot(q)
```
