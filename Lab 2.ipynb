```python
import gensim.downloader as api
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
model = api.load("glove-wiki-gigaword-100")
```

```python
words = [
    "computer",
    "internet",
    "software",
    "hardware",
    "keyboard",
    "mouse",
    "server",
    "network",
    "programming",
    "database",
]

vectors = [model[word] for word in words]
pca = PCA(2).fit_transform(vectors)
```

```python
print(f"Top 5 words similar to 'computer':\n")
similar_words = model.most_similar(positive=["computer"], topn=5)

for word, score in similar_words:
    print(f"{word}: {score:.4f}")
```

```python
for i, word in enumerate(words):
    plt.scatter(pca[i, 0], pca[i, 1])
    plt.annotate(word, (pca[i, 0], pca[i, 1]))

plt.title("PCA Visualization of Technology Word Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid()
plt.show()
```
